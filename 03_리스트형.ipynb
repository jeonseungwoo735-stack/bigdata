```python
#  리스트는 어떻게 만들까?
#  형식

a = []
b = [1, 2, 3]
c = ["Life", "is", "too", "short"]
d = [1, 2, "Life", "is"]
e = [1, 2, ["Life", "is"]]

print(type(a))
print(len(a))
print(len(b))
print(len(c))
print(len(d))
print(len(e))
```
