```python
# 객체에 연산할 숫자 지정하기
class Fourcal:
    def setdata(self , first , second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
cal3 = Fourcal()
print(type(cal3))
cal3.setdata(4 , 2)

print(cal3.first)
print(cal3.second)
print(cal3.add())
```

```python
class Fourcal:
    def setdata(self , first , second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
cal7 = Fourcal()
cal7.setdata(10 , 5)

print(cal7.first)
print(cal7.second)
print(cal7.add())
print(cal7.sub())
print(cal7.mul())
print(cal7.div())
```
