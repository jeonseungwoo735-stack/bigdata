```python
a = [1, 2, 3]
print(id(a))
b = [1, 2, 3]
print(id(b))

a.append(4)
print(a)
print(id(a))
```

```python
a = [1, 2, 3]
b = a[:]
print(id(a))
print(id(b))
```

```python
# 변수를 만드는 여러가지 방법
a , b = ["Python" , "like"]
print(type(a))
print(a)
print(b)
print(id(a))
print(id(b))
```

```python
a , b = "Python" , "like"
print(a)
print(b)
print(type(a))
```

```python
a = b = "Python"
print(a)
print(b)
print(type(a))
print(id(a))
print(id(b))
```
