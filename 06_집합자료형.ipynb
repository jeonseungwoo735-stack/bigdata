```python
# 집합 자료형은 어떻게 만들까?
a = set([1, 2, 3])
print(type(a))
print(a)
```

```python
# 교집합
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
print(s1.intersection(s2))

print(s1 & s2)
```

```python
# 합집합
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

print(s1.union(s2))
print(s1 | s2)
```

```python
# 차집합
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

print(s1.difference(s2))
print(s2.difference(s1))
print(s1 - s2)
print(s2 - s1)
```

```python
# 값 1개 추가하기
s1 = set([1, 2, 3])
s1.add(4)
print(s1)
```

```python
# 값을 여러개 추가하기 update
s1 = set([1, 2, 3])
s1.update([4, 5, 6])
print(s1)
```

```python
# 특정 값 제거하기 remove
s1 = set([1, 2, 3])
s1.remove(2)
print(s1)
```

```python
# 모든 값 제거하기 
s1 = set([1, 2, 3])
s1.clear()
print(s1)
```
