```python
num = int(input("숫자를 입력하세요: "))

if num <= 0:
    print("양수만 입력 가능합니다.")
else:
    i = 1
    total = 0

    while i <= num:
        total += i
        if i < num:
            print(i, end=" + ")
        else:
            print(i, end=" = ")
        i += 1

    print(total)
```

```python
odd = []
even = []

for i in range(1, 21):
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)

print("홀수:", end=" ")
for i in odd:
    print(i, end=", ")

print("(총", len(odd), "개)")

print("짝수:", end=" ")
for i in even:
    print(i, end=", ")

print("(총", len(even),"개)")
```

```python
height = int(input("높이를 입력하세요: "))

if height < 1 or height > 10:
    print("1~10 사이의 숫자만 입력 가능합니다.")
else:
    for i in range(1, height + 1):

        print(" " * (height - i), end = "")

        for j in range(1, i + 1):
            print(j, end="")

        for j in range(i - 1, 0, -1):
            print(j, end="")

        print()
```

```python
num = int(input("숫자를 입력하세요: "))

if num < 2:
    print("2 이상의 숫자만 입력 가능합니다.")
else:
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False

    if prime:
        print(num, "은(는) 소수입니다.")
    else:
        print(num, "은(는) 소수가 아닙니다.")

    print("\n1부터", num, "까지의 소수:")

    count = 0
    for i in range(2, num + 1):
        prime = True
        for j in range(2, i):
            if i % j == 0:
                prime = False

        if prime:
            print(i, end=", ")
            count += 1

    print("\n총", count, "개의 소수가 있습니다.")
```

```python
balance = 10000

while True:
    print("=== ATM 시스템 ==="), 
    print("1. 잔액 조회")
    print("2. 입금")
    print("3. 출금")
    print("4. 종료")

    choice = int(input("선택: "))

    if choice == 1:
        print("현재 잔액:", balance, "원")

    elif choice == 2:
        money = int(input("입금액: "))
        balance += money
        print("입금 완료! 현재 잔액:", balance, "원")

    elif choice == 3:
        money = int(input("출금액: "))
        if money > balance:
            print("잔액이 부족합니다. 현재 잔액:", balance, "원")
        else:
            balance -= money
            print("출금 완료! 현재 잔액:", balance, "원")

    elif choice == 4:
        print("이용해 주셔서 감사합니다.")
        break

    else:
        print("잘못된 선택입니다.")
```

```python
import random

answer = random.randint(1, 100)
count = 0

print("1부터 100사이의 숫자를 맞춰보세요!")

while True:
    guess = int(input("숫자를 입력하세요: "))
    count += 1

    if guess > answer:
        print("Down!")
    elif guess < answer:
        print("Up!")
    else:
        print("정답입니다!", count, "번 만에 맞추셨습니다!")
        break
```

```python
year = int(input("년도: "))
month = int(input("월: "))

if month < 1 or month > 12:
    print("1~12월만 입력 가능합니다.")
else:
    days = [31,28,31,30,31,30,31,31,30,31,30,31]

    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        days[1] = 29

    day = days[month-1]

    print()
    print("    ",f"{year}년", month, "월")

    count = 0
    for i in range(1, day + 1):
        print(f"{i:2}", end=" ")
        count += 1

        if count % 7 == 0:
            print()
```

```python
# 입력값이 몇 개가 될지 모를때는 어떻게 할까?
def add_many(*args):
    result = 0
    print(type(args))
    print(args)

    for i in args:
        result += i
    return result 

result = add_many(1, 2, 3, 4, 5)
print(result)
```

```python
def add_mul(choice , *args):
    if choice == "add":
        result = 0
        for i in args:
            result += i
    
    elif choice == "mul":
        result = i
        for i in args:
            result *= i
    
    return result 
```

```python
result = add_mul("add" , 1 , 2 , 3 , 4 , 5)
print(result)
```

```python
result = add_mul("mul" , 1 , 2 , 3 , 4 , 5)
print(result)
```

```python
# 매개변수에 초기값 미리 설정하기

def say_myself(name , age , man=True):
    print(f"너의 이름은 {name}이다.")
    print(f"너의 나이는 {age}이다.")
    if man:
        print("남자입니다.")
    else:
        print("여자입니다.")

say_myself("홍길동" , 25)
say_myself("김영희" , 30 , False)
```

```python
# 함수 안에서 선언한 변수의 효력 범위

a = 1
def vartest(a):
    a += 1

vartest(a)
print(a)
```

```python
a = 1
def vartest():
    global a
    a += 1

vartest()
print(a)
```

```python
# lambda 예약어
# 함수이름= lambda 매개변수1 , 매개변수2 , ... : 수행할 문장
def add(a , b):
    return a + b

add = lambda a , b : a + b
```

```python
# 함수의 독스트링(Docstring)
def add(a , b):
    """
    두 숫자를 더하는 함수

    Parameters:
    a(int , float): 첫 번째 숫자
    b(int , float): 두 번째 숫자

    retunrs:
    int , float : 두 숫자의 합

    """
    return a + b
print(add.__doc__)
```
