```python
# 튜플은 어떻게 만들까?
a = ()
b = (1,)
c = (1, 2, 3)
# 하나인 요소일때만 뒤에 콤마
d = 1, 2, 3
e = ("a", "b", "ab", "cd")
f = 1,
g = 1
print(type(a)) 
print(type(b)) 
print(type(c)) 
print(type(d)) 
print(type(e)) 
print(type(f)) 
print(type(g)) 
```

```python
# 튜플 요소값을 삭제하여 할 때
a = (1, 2, "a", "b")
del a[0]
```

```python
# 튜플 요소값을 변경하려 할 때
a = (1, 2, "a", "b")
a[0] = "c"
```

```python
a = (1, 2)
a = a * 2
print(a)
```
